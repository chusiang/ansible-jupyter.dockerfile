### Ad-Hoc commands

ping the localhost.

```python
!ansible localhost -m ping
```

Get the facts with `setup` module.

```python
!ansible localhost -m setup
```

Remove the **vim** with **zypper** package management on **openSUSE**.

```python
!ansible localhost -m zypper -a 'name=vim state=absent'
```

Install the **vim** with **zypper** package management on **openSUSE**.

```python
!ansible localhost -m zypper -a 'name=vim state=present'
```

Install the **tree** with **zypper** package management on **openSUSE**.

```python
!ansible localhost -m zypper -a 'name=tree state=present'
```
